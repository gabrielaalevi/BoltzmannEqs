```python
import numpy as np
import matplotlib.pyplot as plt


try:
    import mplhep as hep
    hep.style.use("LHCb2")
except:
    print("mplhep not found. Falling back to default plot style")

```

```python
solutionsFile = './solution_Coann.csv'
```

```python
data = np.genfromtxt(solutionsFile,delimiter=',',names=True)
print(data.dtype.names)
omh2 = 0.12*(data['Yxd'][-1]+data['Yb1'][-1])/(6.8e-13)
print(f'Omega*h^2 = {omh2:1.4g}')
print(len(data['x']))
```

```python
# fig = plt.figure(figsize=(10,8))
plt.plot(data['x'],data['Yxd'], 'r', linewidth=2, label = r'$Y_{\chi}$')
plt.plot(data['x'],data['Yb1'], 'b', linewidth=2, label = r'$Y_{\chi}$',marker='o')

plt.xlabel(r'$x = m_{\chi}/T$')
plt.ylabel('Yield')

plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.show()
```
